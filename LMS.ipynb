### Executive Summery

The purpose of this report is to demonstrate a project proposal and plan for a learning management system (LMS). The proposed LMS is to serve as a portal for students and teachers to conduct online and virtual courses without the need of external support and system. Due to covid-19, the market and importance of online learning management systems has exponentially increased over the last year. On the request from our client, we have conducted a review on current LMSs available in the market and identified few common weaknesses among those systems and have come with features and functions as a solution for the proposed LMS. These features include but not limited to gamification of user interface (UI), intelligent group management, integrated communication and much more. We believe these new features will elevate the user experience (UX) of the proposed system and will stand out as unique and one stop solution for the customers. This report demonstrates this proposed LMS by utilizing various unified modelling language (UML) diagrams and models for easy to understand and standard approach. The proposal also contains set of business rules set out by the client and legislative requirements and a project charter to follow up with the work and methodology of the project team. This report is limited to be an initial documentation which means it lacks the depth of information required for the development of this project, so a second iteration of this report is needed to showcase an in-depth proposal.
